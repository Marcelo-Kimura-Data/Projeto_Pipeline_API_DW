```python
import pandas as pd
```

```python
df_silver = pd.read_csv(f"2_bronze/bronze_cripto_180d.csv")
```

```python
df_silver
```

```python
df_silver.columns
```

```python
df_silver.drop(columns=['code', 'codein','name','create_date'], inplace=True)
```

```python
colunas_renomeadas = {
    'high': 'preco_maximo',
    'low': 'preco_minimo',
    'varBid': 'variacao_bid',
    'pctChange': 'variacao_pct',
    'bid': 'preco_compra',
    'ask': 'preco_venda',
    'timestamp': 'data_hora',
    'moeda': 'moeda'
}
```

```python
df_silver.rename(columns=colunas_renomeadas, inplace=True)
```

```python
df_silver['moeda'] = df_silver['moeda'] + '-BRL'
```

```python
df_silver.head(5)
```

```python
df_silver.info()
```

```python
df_silver['data_hora'] = pd.to_datetime(df_silver['data_hora'], unit='s')
df_silver['data_hora'] = df_silver['data_hora'].dt.strftime('%d/%m/%Y %H:%M')
```
