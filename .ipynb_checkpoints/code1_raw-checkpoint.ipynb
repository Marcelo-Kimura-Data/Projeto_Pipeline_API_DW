```python
import requests as re
import os
import json
import time
import pprint
```

```python
criptos = ['BTC','ETH','LTC','XRP','DOGE','SOL']
dias = '180'
dados_criptos = {}
```

```python
for moeda in criptos:
    url = f'https://economia.awesomeapi.com.br/json/daily/{moeda}-BRL/{dias}'
    print(f"[INFO] Extraindo dados da moeda: {moeda} ({dias} dias)")

    response = re.get(url)

    if response.status_code == 200:
        data = response.json()
        dados_criptos[moeda] = data
        print(f"[OK] {moeda}: {len(data)} registros extraídos.")
    else:
        print(f"[ERRO] Falha ao extrair {moeda}. Status code: {response.status_code}")
    
    time.sleep(5)  # dá uma segurada para não spammar a API
```

```python
# Opcional
pprint.pprint(dados_criptos)
```

```python
filename = f"1_raw/dados_cripto_{dias}d.json"

with open(filename, 'w') as f:
    json.dump(dados_criptos, f, indent=4)
```
