```python
import pandas as pd
```

```python
df_silver = pd.read_csv(f"2_bronze/bronze_cripto_180d.csv")
```

```python
df_silver.info()
df_silver.head(5)
```

```python
df_silver.columns
```

```python
# Excluir colunas que não serão utilizadas:

df_silver.drop(columns=['code', 'codein','name','create_date'], inplace=True)
```

```python
# renomeando as colunas para ficar mais claro as informações:

colunas_renomeadas = {
    'high': 'preco_maximo',
    'low': 'preco_minimo',
    'varBid': 'variacao_bid',
    'pctChange': 'variacao_pct',
    'bid': 'preco_compra',
    'ask': 'preco_venda',
    'timestamp': 'data_hora',
    'moeda': 'moeda'
}
df_silver.rename(columns=colunas_renomeadas, inplace=True)
```

```python
# Adicionando a nomenclatura para ficar mais claro o tipo de cotação (código Cripto + código real):

df_silver['moeda'] = df_silver['moeda'] + '-BRL'
```

```python
# Conversão do formato de data e hora da coluna 'data_hora' (segundos -> data + hora

df_silver['data_hora'] = pd.to_datetime(df_silver['data_hora'], unit='s')
df_silver['data_hora'] = df_silver['data_hora'].dt.strftime('%d/%m/%Y %H:%M')
```

```python
df_silver.info()
df_silver.head(10)
```

```python
# Salvando arquivo na camada silver

df_silver.to_csv("3_silver/silver_cripto_180d.csv", index=False)
```
