```python
import json
import pandas as pd
```

```python
filename = "1_raw/dados_cripto_180d.json"
dados_criptos = {}

with open(filename, 'r') as f:
    dados_criptos = json.load(f)
```

```python
dados_criptos
```

```python
lista_registros = []

for moeda, registros in dados_criptos.items():
    for registro in registros:
        registro['moeda'] = moeda  # adiciona coluna com o nome da moeda
        lista_registros.append(registro)
```

```python
df_bronze = pd.DataFrame(lista_registros)
```

```python
df_bronze.head(10)
```

```python
df_bronze.to_csv("2_bronze/bronze_cripto_180d.csv", index=False)
```
